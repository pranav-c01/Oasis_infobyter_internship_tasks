### Data Cleaning And Transformation

#### First of all drop Car_Name cols , as it has very low correlation with the Selling Price column

```python
df.drop('Car_Name',axis=1,inplace=True)
```

### Remove Outliers from Numerical Vars

```python
def remove_outliers_iqr(data, cols):
    for col in cols:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        data = data[data[col] >= lower_bound]
        data = data[data[col] <= upper_bound]
    return data

# Remove outliers using IQR
data = remove_outliers_iqr(df.copy(), numerical_cols[1:])
data.index = [x for x in range(0,276)]
```

### Numerical Features sqrt Transformation to make them Standard Distribution

```python
numerical_cols
```

```python
for i in numerical_cols[2:]:
    data[i] = np.power(data[i],0.5)
```

## Categorical Columns Ordinal Encoder transform

```python
categorical_cols[1:]+["Year"]
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Label encode categorical columns
le = OrdinalEncoder()
le.fit(data[categorical_cols[1:-1]+["Year"]])
data_cat = le.transform(data[categorical_cols[1:-1]+["Year"]])
data_cat
```

```python
le.feature_names_in_
```

```python
data_final = pd.concat([data.drop(columns=categorical_cols[1:-1]+["Year"]),pd.DataFrame(data_cat,columns=le.feature_names_in_)],axis=1)
data_final.head()
```

## Model Selection

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Ensure correct dataframe structure (numerical and categorical features)
# ... (your code to handle categorical encoding and numerical transformation)

# Combine numerical and categorical features into a single dataframe
# features = pd.concat([numerical_df, categorical_df], axis=1)  # Update with your dataframe names

# Split data into training and testing sets
X,y = data_final.drop('Selling_Price',axis=1),data_final['Selling_Price']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42)

# Define candidate models and hyperparameter grids
models = {
    "LinearRegression": {
        "fit_intercept": [True, False],
        "normalize": [True, False],
    },
    "Ridge": {
        "alpha": [0.1, 1, 10],
    },
    "Lasso": {
        "alpha": [0.1, 1, 10],
    },
    "RandomForestRegressor": {
        "n_estimators": [100, 200, 400],
        "max_depth": [3, 5, 7],
    },
    "GradientBoostingRegressor": {
        "n_estimators": [100, 200, 400],
        "max_depth": [3, 5, 7],
        "learning_rate": [0.1, 0.01, 0.001],
    },
    "XGBRegressor": {
        "n_estimators": [100, 200, 400],
        "max_depth": [3, 5, 7],
        "learning_rate": [0.1, 0.01, 0.001],
    },
    "DecisionTreeRegressor": {
        "max_depth": [3, 5, 7],
    },
}

# Train and evaluate each model using GridSearchCV
best_model_name = None
best_model = None
best_score = -float("inf")  # Start with negative infinity for minimum MSE

for model_name, params in models.items():
    model = eval(model_name)()  # Use eval for demonstration, choose appropriate import
    grid_search = GridSearchCV(model, params, cv=5, scoring="neg_mean_squared_error")  # Minimize MSE
    grid_search.fit(X_train, y_train)

    if grid_search.best_score_ > best_score:
        best_model_name = model_name
        best_model = grid_search.best_estimator_
        best_score = grid_search.best_score_

# Make predictions on test data with the best model
y_pred = best_model.predict(X_test)

# Evaluate final model performance on test data
print(f"Best model: {best_model_name} with score: {best_score:.4f} (mean squared error)")
print(f"R-squared: {r2_score(y_test, y_pred):.4f}")

# Consider additional performance metrics and visualizations as needed
# ... (plot actual vs. predicted values, etc.)

# Save the best model and preprocessing objects for future use
# ... (your code to save model and preprocessing artifacts)
```
